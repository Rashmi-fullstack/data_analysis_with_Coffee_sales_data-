### Question 4: Data Type Conversion (1 Marks)

The 'Date' column is currently an object. Convert it to a datetime object. Then, create two new columns: 'Month' (containing the month number) and 'Year' (containing the year as an integer). Finally, display the data types of all columns.

**Expected Output:** A series listing each column and its updated data type.

```python
coffee_data['Date'] = pd.to_datetime(coffee_data['Date'])
coffee_data['Month_name'] = coffee_data['Date'].dt.month
coffee_data['Year'] = coffee_data['Date'].dt.year
coffee_data.dtypes
```

### Question 5: Data Aggregation and Analysis (5 Marks)

Perform the following data analysis tasks:
1. **Average Sales per Year:** Calculate the average 'money' (sales) for each 'Year'. Group the data by the 'Year' column and then compute the mean of the 'money' column for each group.
2. **Maximum Sales by Month Name:** Find the maximum 'money' earned for each 'Month_name'. Display the result as a Pandas Series.
3. **Total Sales by Coffee Name:** Calculate the total 'money' (sum of sales) for each 'coffee_name'.
4. **Unique Coffee Types:** Determine how many unique types of coffee are present in the dataset. Print the unique coffee names as a list.
5. **Average Sales by Time of Day:** Find the average 'money' made during each 'Time_of_Day' (Morning, Afternoon, Night).

**Expected Output:** Display the appropriate results for each step (e.g. Pandas Series and list) clearly.

```python
avg_sales_year = coffee_data.groupby('Year')['money'].mean()
print("Average sales per year:")
print(avg_sales_year)

max_sales_month = coffee_data.groupby('Month_name')['money'].max()
print("\nMaximum sales in each month:")
print(max_sales_month)

total_sales_coffee = coffee_data.groupby('coffee_name')['money'].sum()
print("\nTotal sales by coffee name:")
print(total_sales_coffee)

unique_coffee_type = coffee_data['coffee_name'].unique()
print("\nCoffee types:")
print(unique_coffee_type)

avg_sales_time = coffee_data.groupby('Time_of_Day')['money'].mean()
print("\nAverage sales separately according to the time of day")
print(avg_sales_time)
```

### Question 7: Load MNIST Data (4 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits
digits = load_digits()
x = digits.data
y = digits.target
print('Shape of x', x.shape)
print('Shape of y', y.shape)
```

### Question 8: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10,random_state=42)
kmeans_labels = kmeans.fit_predict(x)
print('cluster labels:', kmeans_labels)
```

### Question 9: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from scipy.stats import mode
from sklearn.metrics import f1_score

mapped_labels = np.zeros_like(kmeans_labels)

for i in range(10):
    mask = (kmeans_labels == i)
    mapped_labels[mask] = mode(y[mask], keepdims=False).mode

f1 = f1_score(y, mapped_labels, average='macro')
print("Macro F1 Score:", f1)
```

### Question 10: Confusion Matrix for Clustering (3 Marks)

Generate and display the confusion matrix comparing your mapped cluster labels (from Question 9) to the true labels (`y`). Use `confusion_matrix` from `sklearn.metrics`. This helps evaluate how well your clusters align with the actual digit classes without requiring visual plotting.

**Expected Output:** A 10x10 confusion matrix printed as an array or a Pandas DataFrame.

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y, mapped_labels)

cm_df = pd.DataFrame(
    cm,
    index=[f"True_{i}" for i in range(10)],
    columns=[f"Pred_{i}" for i in range(10)]
)

print(cm_df)
```
